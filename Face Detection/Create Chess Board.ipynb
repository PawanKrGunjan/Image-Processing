### 1. Let's Create a Chess Board

```python
zeros = np.zeros((8,8), dtype = int)
zeros
```

```python
for i in range(len(zeros)):
    for j in range(len(zeros[i])):
        if ((i+j)%2)==0:
            zeros[i][j]=1
zeros
```

```python
plt.figure(figsize=(8,8), dpi=100)            
plt.imshow(zeros, cmap=plt.cm.binary) 
plt.show()
```

```python
ones = np.ones((8,8), dtype = int)

for i in range(len(ones)):
    for j in range(len(ones[i])):
        if ((i+j)%2)==0:
            ones[i][j]=0
            
plt.figure(figsize=(8,8), dpi=100)            
plt.imshow(ones, cmap=plt.cm.binary) 
plt.show()
```

### 2. Let's create a high pixel density Chess image of size (1280 X 1280) pixel

```python
img = np.ones((1280,1280), dtype = int)
zeros = np.zeros((160,160), dtype = int)

for i in range(1,9):
    for j in range(1,9):
        if ((i+j)%2)==0:
            img[(i-1)*160:i*160,(j-1)*160:j*160] = zeros
img
```

```python
plt.figure(figsize=(8,8), dpi=100)            
plt.imshow(img, cmap=plt.cm.binary) 
plt.show()
```
